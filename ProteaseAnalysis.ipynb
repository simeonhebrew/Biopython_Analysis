```python
#Reading through the sequence file
novelcov_record = SeqIO.read("../Data/sarsseq.fasta", "fasta")
novelcov_record
```

```python
#Viewing Sequence
novelcov_dna = novelcov_record.seq
novelcov_dna
```

```python
#Getting length of the sequence
len(novelcov_dna)
```

```python
#Transcribing the nucleotide sequence
novelcov_mrna = novelcov_dna.transcribe()
novelcov_mrna 
```

```python
#Translating mRNA to proteins
novelcov_proteins = novelcov_mrna.translate()
novelcov_proteins
```

```python
#Calculating length of protein
len(novelcov_proteins)
```

```python
#Estimating length of protein
len(novelcov_dna)/3
```

```python
#Showing sequences before stop codon

novelcov_aa = novelcov_proteins.split("*")
novelcov_aa
```

```python
novelcov_clean = [str(i) for i in novelcov_aa]
novelcov_clean
```

```python
#Extracting amino acids ranges (before stop codon)

import pandas as pd
protdf = pd.DataFrame({'Amino acids': novelcov_clean})
protdf
```

```python
#Counting length of amino acids
protdf['count'] = protdf['Amino acids'].str.len()
protdf.head()
```

```python
#The longest amino acid value
protdf.nlargest(10,"count")
```

```python
#Calculating the frequency of amino acids

from collections import Counter

Counter(novelcov_proteins).most_common(10)
```
